```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats

from sklearn.preprocessing import StandardScaler, MinMaxScaler
dataset = pd.read_csv('titanic.csv')
dataset
```

```python
fare_values = dataset.loc[dataset['Fare'].notnull(), 'Fare'].values
print(fare_values.shape)
```

```python
plt.hist(fare_values, bins=25)
plt.show()
```

#### Zdecydowanie najwięcej biletów było w cenie miedzy 1-40, natomiast najczesciej oplata wynosila miedzy 1-20.

```python
mean_fare = np.round(np.mean(fare_values))
print(mean_fare)
```

#### Srednia cena biletu to 32.0

```python
plt.hist(fare_values, bins=25)
plt.axvline(x = mean_fare, color='red', label = 'Średnia')
plt.legend(loc='upper right')
plt.show()
```

```python
median_fare = np.round(np.median(fare_values))
print(median_fare)
```

#### Mediana ceny biletu to 14.0

```python
plt.hist(fare_values, bins=25)
plt.axvline(x = mean_fare, color='red', label = 'Średnia')
plt.axvline(x = median_fare, color='green', label = 'Mediana')
plt.legend(loc='upper right')
plt.show()
```

```python
mode_fare = stats.mode(fare_values)
print(mode_fare)
```

#### Najczesciej wystepujaca wartosc oplaty za bilet wynosi 8.05 i takich biletow zakupiono 43 sztuki

```python
plt.hist(fare_values, bins=20)
plt.axvline(x = mean_fare, color='red', label = 'Średnia')
plt.axvline(x = median_fare, color='green', label = 'Mediana')
plt.axvline(x = mode_fare[0], color='orange', label = 'Moda')
plt.legend(loc='upper right')
plt.show()
```

```python
q0 = np.quantile(fare_values, 0.0)
q1 = np.round(np.quantile(fare_values, 0.25))
q3 = np.round(np.quantile(fare_values, 0.75))
q4 = np.round(np.quantile(fare_values, 1.0))
print(f'Q0: {q0}')
print(f'Q1: {q1}')
print(f'Q2: {median_fare}')
print(f'Q3: {q3}')
print(f'Q4: {q4}')
```

#### Najnizsza cena biletu wynosila 0.0 (darmowy bilet), natomiast najwyzsza cena biletu to 512.0. 75% wszystkich biletow nie przekroczylo ceny 31.0. Widac tutaj zdecydowana roznice pomiedzy najdrozszym biletem a pozostalym cenami biletow.

```python
plt.hist(fare_values, bins=25)
plt.axvline(x = mean_fare, color='red', label = 'Średnia')
plt.axvline(x = median_fare, color='green', label = 'Mediana')
plt.axvline(x = mode_fare[0], color='orange', label = 'Moda')
plt.axvline(x = q0, color='black', label = 'Min')
plt.axvline(x = q1, color='black', label = 'Q1')
plt.axvline(x = q3, color='black', label = 'Q3')
plt.axvline(x = q4, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

```python
range_fare = round((max(fare_values) - min(fare_values)), 2)
print(range_fare)
```

#### Zakres cen wynosi 512.33

```python
iqr = q3 - q1
print(iqr)
```

#### Jest wiele wartosci odstajacych dlae cen powyzej 100, co swiadczy o tym, ze w tym miejcu wartosci sa rzadkie.

```python
variance_fare = np.var(fare_values, ddof=1)
print(variance_fare)
```

```python
print(variance_fare**(1/2))
```

```python
standard_deviation_fare = np.std(fare_values, ddof=1)
print(round(standard_deviation_fare,2))
```
