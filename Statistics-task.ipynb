```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats

from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
data_100 = np.random.normal(loc=3, scale=1, size=100)
```

```python
data_10000 = np.random.normal(loc=3, scale=1, size=10000)
```

## Dane z rozkładu normalnego o średniej 3 i odchyleniu standardowym 1 dla 100

```python
print("Rozmiar probki:", data_100.shape)
```

```python
plt.hist(data_100, bins=25)
plt.show()
```

```python
mean_data_100 = np.round(np.mean(data_100),2)
print("Średnia:", mean_data_100)
plt.hist(data_100, bins=25)
plt.axvline(x=mean_data_100, color='red', label='Średnia')
plt.legend(loc='upper right')
plt.show()
```

```python
median_data_100 = np.round(np.median(data_100),2)
print("Mediana:", median_data_100)
plt.hist(data_100, bins=25)
plt.axvline(x=mean_data_100, color='red', label='Średnia')
plt.axvline(x=median_data_100, color='green', label='Mediana')
plt.legend(loc='upper right')
plt.show()
```

```python
q0 = np.quantile(data_100, 0.0)
q1 = np.round(np.quantile(data_100, 0.25))
q3 = np.round(np.quantile(data_100, 0.75))
q4 = np.round(np.quantile(data_100, 1.0))
print(f'Q0: {q0}')
print(f'Q1: {q1}')
print(f'Q2: {median_data_100}')
print(f'Q3: {q3}')
print(f'Q4: {q4}')
```

```python
plt.hist(data_100, bins=25)
plt.axvline(x = mean_data_100, color='red', label = 'Średnia')
plt.axvline(x = median_data_100, color='green', label = 'Mediana')
plt.axvline(x = q0, color='black', label = 'Min')
plt.axvline(x = q1, color='black', label = 'Q1')
plt.axvline(x = q3, color='black', label = 'Q3')
plt.axvline(x = q4, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

```python
range_data_100 = round((max(data_100) - min(data_100)), 2)
print(range_data_100)
```

```python
iqr = q3 - q1
print(iqr)
```

```python
plt.boxplot(data_100)
plt.show()
```

```python
variance_data_100 = np.var(data_100, ddof=1)
print(variance_data_100)
print(variance_data_100**(1/2))
standard_deviation_data_100 = np.std(data_100, ddof=1)
print(round(standard_deviation_data_100,2))
```

### Komentarz: 
- Średnia (3.18) i mediana (3.16) są do siebie bardzo zbliżone, co wskazuje na prawie symetryczny rozkład danych.
- Wartości są skoncentrowane wokół średniej 3, a niewielkie odchylenie wynika z losowości próby o wielkości 100 obserwacji.
- Połowa danych mieści się w przedziale od 2 do 4 (IQR = 2), co oznacza umiarkowane rozproszenie.
- Odchylenie standardowe wynosi około 1.02, czyli jest bardzo bliskie wartości teoretycznej równej 1.
