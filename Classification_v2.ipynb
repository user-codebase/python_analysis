```python
import matplotlib.pyplot as plt
from numpy import where, meshgrid, arange, hstack

def plot_classification_surface_multiclass(X_plot, y_plot, trained_model):
    plt.figure(figsize=(12, 7))

    # granice osi
    min1, max1 = X_plot[:, 0].min()-1, X_plot[:, 0].max()+1
    min2, max2 = X_plot[:, 1].min()-1, X_plot[:, 1].max()+1

    # siatka punktów
    x1grid = arange(min1, max1, 0.1)
    x2grid = arange(min2, max2, 0.1)
    xx, yy = meshgrid(x1grid, x2grid)

    # przetworzenie siatki w wektor
    r1, r2 = xx.flatten(), yy.flatten()
    r1, r2 = r1.reshape((len(r1),1)), r2.reshape((len(r2),1))
    grid = hstack((r1,r2))

    # predykcja dla każdego punktu siatki
    yhat = trained_model.predict(grid)
    zz = yhat.reshape(xx.shape)

    # kontur dla klas
    plt.contourf(xx, yy, zz, cmap='Paired', alpha=0.3)

    # punkty treningowe
    colors = ['green', 'red', 'blue']  # dla 3 gatunków
    for class_value, color in zip(range(3), colors):
        row_ix = where(y_plot == class_value)
        plt.scatter(X_plot[row_ix, 0], X_plot[row_ix, 1], color=color, alpha=0.7, label=f'Species {class_value}')

    plt.xlabel('CulmenLength')
    plt.ylabel('FlipperLength')
    plt.title('Granica decyzyjna - Regresja logistyczna (penguins)')
    plt.legend(loc='upper right')
    plt.show()

# przykład dla dwóch cech: CulmenLength i FlipperLength
plot_classification_surface_multiclass(
    X_plot=X_train_standardized, 
    y_plot=y_train.values, 
    trained_model=model_lr
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, f1_score, roc_curve, roc_auc_score
from sklearn.preprocessing import label_binarize

metrics_dataframe = pd.DataFrame(columns = ['Model', 'F1_score', 'AUC'])
models = []
models_names = []
predictions_proba_list = []

def calculate_metrics(model, name, X_checked, y_checked):
    models.append(model)
    models_names.append(name)
    global metrics_dataframe
    
    # Predykcje
    predictions = model.predict(X_checked)
    predictions_proba = model.predict_proba(X_checked)
    predictions_proba_list.append(predictions_proba)
    
    ############## metryki dla sprawdzanego modelu ################
    print(f"\nClassification Report for {name}:\n")
    print(classification_report(y_checked, predictions))
    
    # Confusion matrix
    plt.figure()
    cm = confusion_matrix(y_checked, predictions)
    ax = sns.heatmap(cm, annot=True, cmap='Blues', fmt='.0f')
    ax.set_title(f'Confusion Matrix - {name}\n\n')
    ax.set_xlabel('\nPredicted Values')
    ax.set_ylabel('Actual Values ')
    plt.show()
    
    # ROC Curve dla każdej klasy
    y_checked_bin = label_binarize(y_checked, classes=[0,1,2])  # dla 3 klas
    n_classes = y_checked_bin.shape[1]
    
    plt.figure(figsize=(6,6))
    plt.plot([0,1],[0,1],'k--')  # linia losowego modelu
    for i in range(n_classes):
        fpr, tpr, _ = roc_curve(y_checked_bin[:, i], predictions_proba[:, i])
        auc_class = roc_auc_score(y_checked_bin[:, i], predictions_proba[:, i])
        plt.plot(fpr, tpr, label=f'Class {i} (AUC = {auc_class:.2f})')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - {name} (One-vs-Rest)')
    plt.legend(loc='lower right')
    plt.show()
    
    # Metryki F1 i AUC dla całego problemu wieloklasowego
    f1_metric = f1_score(y_checked, predictions, average='macro')
    auc_metric = roc_auc_score(y_checked_bin, predictions_proba, average='macro')
    
    # nowy wiersz w DataFrame przez pd.concat
    new_row = pd.DataFrame([{'Model': name, 'F1_score': f1_metric, 'AUC': auc_metric}])
    metrics_dataframe = pd.concat([metrics_dataframe, new_row], ignore_index=True)
    
    return metrics_dataframe
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.multiclass import OneVsRestClassifier

# Funkcja wizualizująca granice decyzyjne w 2D
def plot_classification_surface(X_plot, y_plot, trained_model, feature_names=None):
    plt.figure(figsize=(12, 7))
    
    # granice osi
    min1, max1 = X_plot[:, 0].min()-1, X_plot[:, 0].max()+1
    min2, max2 = X_plot[:, 1].min()-1, X_plot[:, 1].max()+1
    
    x1grid = np.arange(min1, max1, 0.1)
    x2grid = np.arange(min2, max2, 0.1)
    xx, yy = np.meshgrid(x1grid, x2grid)
    
    r1, r2 = xx.flatten().reshape(-1,1), yy.flatten().reshape(-1,1)
    grid = np.hstack((r1,r2))
    
    # predykcja
    yhat = trained_model.predict(grid)
    zz = yhat.reshape(xx.shape)
    
    plt.contourf(xx, yy, zz, alpha=0.3, cmap='Set1')
    
    classes = np.unique(y_plot)
    colors = ['red', 'green', 'blue']
    for class_value, color in zip(classes, colors):
        idx = np.where(y_plot == class_value)
        plt.scatter(X_plot[idx,0], X_plot[idx,1], c=color, label=f'Class {class_value}', edgecolor='k', s=80)
    
    plt.xlabel(feature_names[0] if feature_names else 'Feature 1')
    plt.ylabel(feature_names[1] if feature_names else 'Feature 2')
    plt.title('Decision Surface for Multi-class Classification')
    plt.legend(loc='upper left')
    plt.show()


# zakładamy, że X_train_standardized i X_test_standardized zawierają już tylko 2 cechy
features_to_plot = ['CulmenLength', 'FlipperLength']

# nie wybieramy indeksów, tylko używamy całej macierzy
X_train_2feat = X_train_standardized
X_test_2feat  = X_test_standardized

# Różne regularyzacje i wartości C
penalties = ['l1', 'l2']
C_values = [0.01, 0.1, 1, 10]

results = []

for penalty in penalties:
    for C in C_values:
        print("=================================")
        print(f"Penalty: {penalty}, C: {C}")
        
        # model wieloklasowy One-vs-Rest
        model = OneVsRestClassifier(
            LogisticRegression(
                penalty=penalty,
                C=C,
                solver='liblinear',  # liblinear obsługuje l1 i l2
                max_iter=1000
            )
        )
        
        # trenowanie
        model.fit(X_train_2feat, y_train)
        
        # predykcja
        train_pred = model.predict(X_train_2feat)
        test_pred = model.predict(X_test_2feat)
        
        # F1 score dla problemu wieloklasowego
        f1_train = f1_score(y_train, train_pred, average='macro')
        f1_test = f1_score(y_test, test_pred, average='macro')
        
        print("F1-score TRAIN:", round(f1_train,3))
        print("F1-score TEST :", round(f1_test,3))
        
        # zapis wyników
        results.append({
            "penalty": penalty,
            "C": C,
            "F1_train": f1_train,
            "F1_test": f1_test
        })
        
        # wizualizacja granicy klasyfikacji
        plot_classification_surface(
            X_plot=X_train_2feat,
            y_plot=y_train,
            trained_model=model,
            feature_names=features_to_plot
        )
```

Zmiana rodzaju regularyzacji (L1, L2) oraz parametru C znacząco wpływała na wyniki modelu, zwłaszcza dla małych wartości C (0.01). Przy bardzo małym C (0.01) F1-score był niski (0.48), co oznacza, że model był zbyt regularizowany i nie był w stanie dobrze dopasować się do danych.

Dla większych wartości C (0.1–10) F1-score wzrasta do wysokich wartości (~0.93–0.95 zarówno na zbiorze treningowym, jak i testowym), co wskazuje, że model dobrze nauczył się rozróżniać gatunki pingwinów i brak jest oznak przeuczenia.

Najwyższe wartości F1-score uzyskano przy C = 1 i C = 10 dla obu typów regularyzacji, co sugeruje, że odpowiednio umiarkowana lub słaba regularyzacja jest optymalna dla klasyfikacji gatunków pingwinów przy użyciu dwóch cech (CulmenLength i FlipperLength).
