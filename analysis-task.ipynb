```python
df['DOB'] = pd.to_datetime(df['DOB'],format='%m/%d/%y')
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'],format='%m/%d/%y')
df['DateofHire'] = pd.to_datetime(df['DateofHire'],format='%m/%d/%Y')
```

```python
df = df[df['DateofHire'].notna()]
```

```python
today = pd.Timestamp('today')
df['EndDate'] = df['DateofTermination'].fillna(today)
df['Seniority'] = (df['EndDate'] - df['DateofHire']).dt.days / 365.25
df[['DateofHire', 'DateofTermination', 'EndDate', 'Seniority']].head()
```

```python
seniority_by_source = df.groupby('RecruitmentSource')['Seniority'].agg(
    count='count',
    avg_seniority='mean',
    median_seniority='median'
).sort_values(by='avg_seniority', ascending=False)

seniority_by_source = seniority_by_source[seniority_by_source['count'] >= 5]

seniority_by_source
```

```python
plt.figure(figsize=(10,8))

sns.barplot(
    data=df,
    y='RecruitmentSource',
    x='Seniority',
    estimator='mean',
)

plt.xlabel('Średni staż (lata)')
plt.ylabel('Źródło rekrutacji')
plt.title('Średni staż pracowników w zależności od źródła rekrutacji')

plt.tight_layout()
plt.show()
```

#### Biorac po uwage zrodlo pozyskania pracownika oraz staz pracownika, najlepiej jest pozyskiwac pracownikow od Careerbuilder, On-Campus Recruiting, Website Banner Ads.

```python
current_employees = df.loc[df['DateofTermination'].isna()].copy()

today = pd.Timestamp('today')

current_employees['Age'] = (today - current_employees['DOB']).dt.days / 365
current_employees = current_employees[(current_employees['Age'] >= 15) & (current_employees['Age'] <= 80)]

current_employees[['DOB','Age']]
```

```python
plt.figure(figsize=(10,6))
sns.histplot(current_employees['Age'], bins=15,kde=True)
plt.xlabel('Wiek (lata)')
plt.ylabel('Liczba pracowników')
plt.title('Struktura wieku aktualnie zatrudnionych pracowników')
plt.show()
```

#### Aktualnie najwiecej pracownikow ma wiek zblizony do 39 roku zycia. Zaraz o tej grupie jest duzy odsetek ludzi z wiekiem okolo 46 lat. Analizujac wykres mozna zobaczy, ze jest to organizacja z dojrzalymi ludzmi, gdyz duzo osob jest w granicach 38-50 lat.
